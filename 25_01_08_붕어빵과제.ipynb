```python
class BungeoppangShop:
    def __init__(self):
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
        }
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }
    def check_stock(self):
        print("현재 재고:")
        for bread_type, stock in self.stock.items():
            print(f"{bread_type}: {stock}개")
    def process_order(self, bread_type, bread_count):
        if bread_type not in self.stock:
            print("잘못된 붕어빵 종류입니다.")
            return
        if self.stock[bread_type] < bread_count:
            print(f"{bread_type}의 재고가 부족합니다. 현재 재고: {self.stock[bread_type]}개")
            return
        self.stock[bread_type] -= bread_count
        self.sales[bread_type] += bread_count
        print(f"{bread_type} {bread_count}개 주문이 완료되었습니다.")
    def admin_mode(self):
        while True:
            붕어빵 = input("재고를 추가할 붕어빵 종류를 입력하세요 (팥, 슈크림, 초코): ")
            붕어빵변환 = {
                "팥": "팥붕어빵",
                "슈": "슈크림붕어빵",
                "슈크림": "슈크림붕어빵",
                "초코": "초코붕어빵"
            }
            if 붕어빵 in 붕어빵변환:
                bread_type = 붕어빵변환[붕어빵]
            elif 붕어빵 in self.stock:
                bread_type = 붕어빵
            else:
                print("잘못된 붕어빵 종류입니다. 다시 입력해주세요.")
                continue
            while True:
                try:
                    붕어빵굽기 = int(input("추가할 재고 개수를 입력하세요: "))
                    if 붕어빵굽기 > 0:
                        self.stock[bread_type] += 붕어빵굽기
                        print(f"{bread_type} 재고가 {붕어빵굽기}개 추가되었습니다.")
                        return
                    else:
                        print("재고는 1개 이상 추가할 수 있습니다. 다시 입력해주세요.")
                except ValueError:
                    print("숫자를 입력해주세요.")
    def calculate_total_sales(self):
        total_sales = sum(self.sales[bread_type] * self.prices[bread_type] for bread_type in self.sales)
        print(f"총 매출: {total_sales:,}원")
        return
    def process_order_check(self, bread_type_input, is_admin=False):
        붕어빵변환 = {
            "팥": "팥붕어빵",
            "슈": "슈크림붕어빵",
            "슈크림": "슈크림붕어빵",
            "초코": "초코붕어빵"
        }
        if bread_type_input in 붕어빵변환:
            bread_type = 붕어빵변환[bread_type_input]
        elif bread_type_input in self.stock:
            bread_type = bread_type_input
        else:
            print("잘못된 붕어빵 종류입니다.")
            return None, None
        if is_admin:
            return bread_type
        while True:
            try:
                bread_count = int(input("주문할 개수를 입력하세요: " if not is_admin else "추가할 재고 개수를 입력하세요: "))
                if bread_count > 0:
                    return bread_type, bread_count
                else:
                    print("개수는 1개 이상이어야 합니다. 다시 입력해주세요.")
            except ValueError:
                print("숫자를 입력해주세요.")
                return bread_type, None
def main():
    shop = BungeoppangShop()
    while True:
        print("\n모드를 선택하세요:")
        print("1. 주문")
        print("2. 관리자")
        print("3. 재고")
        print("4. 종료")
        choice = input("선택: ")

        if choice in ["1", "주문"]:
            while True:
                bread_type_input = input("붕어빵 종류를 입력하세요 (팥, 슈크림, 초코): ")
                bread_type, bread_count = shop.process_order_check(bread_type_input)
                if bread_type and bread_count:
                    shop.process_order(bread_type, bread_count)
                    break
        elif choice in ["2", "관리자"]:
            shop.admin_mode()
        elif choice in ["3", "재고"]:
            shop.check_stock()
        elif choice in ["4", "종료"]:
            print("프로그램을 종료합니다.")
            shop.calculate_total_sales()
            break
        else:
            print("잘못된 선택입니다. 다시 시도하세요.")
if __name__ == "__main__":
    main()
```
