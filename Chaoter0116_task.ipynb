__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
종주문내역 = {}
판매내역 = {"팥붕어빵" : 0, "슈크림붕어빵" : 0, "초코붕어빵" : 0}
while True:
    주문 = input("어떤 붕어빵 드릴까요? (뒤에 붕어빵 생략)(팥/슈크림/초코): ")
    if 주문 == "종료":
        break
    주문빵 = 주문 + "붕어빵"
    if 주문빵 not in 붕어빵:
        print("그런 붕어빵은 없는데요..")
        continue
    개수 = int(input("몇 개 드릴까요? "))
    if 붕어빵[주문빵] < 개수:
        print("재고소진이 되어가지고 남은 것중에서 골라주세요")
    else:
        붕어빵[주문빵] -= 개수
        if 주문빵 in 주문내역:
            주문내역[주문빵] += 개수
            판매내역[주문빵] += 개수
        else:
            주문내역[주문빵] = 개수
            판매내역[주문빵] = 개수

            print("주문 내역: ", 주문내역)
            print("남은 재고: ", 붕어빵)

print("남은 재고: ", 붕어빵)
```
